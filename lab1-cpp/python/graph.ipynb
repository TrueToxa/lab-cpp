# Поиск

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Линейный поиск
$\mathcal{O}(N)$

```python
df = pd.read_csv('../csv/find_num_in_array.csv', delimiter=';')

plt.figure(figsize=(10,6))

sns.regplot(data=df, x="n", y="t_ns", order=1, x_estimator=np.mean, ci = 95)
# кресты ошибок - доверительный интервал 95%

# plt.yscale('log')
# plt.xscale('log')

plt.title('Линейный поиск')
plt.xlabel('$N$')
plt.ylabel('$\\tau,\\ \\text{ns}$')
plt.grid(True, alpha=0.3)
# plt.ylim(0, 50)
# plt.xlim(0, 100000)
plt.savefig('find_num_in_array_lin.png')
plt.show()
```

```python
df = pd.read_csv('../csv/find_num_in_array.csv', delimiter=';')

plt.figure(figsize=(10,6))

sns.regplot(x=np.log10(df['n']), y=np.log10(df['t_ns']), order=1, x_estimator=np.mean, ci = 95)
# кресты ошибок - доверительный интервал 95%

# plt.yscale('log')
# plt.xscale('log')

plt.title('Линейный поиск')
plt.xlabel('$\\log_{10}{N}$')
plt.ylabel('$\\log_{10}{(\\tau/\\text{ns})}$')
plt.grid(True, alpha=0.3)
# plt.ylim(0, 50)
# plt.xlim(0, 100000)
plt.savefig('find_num_in_array_log.png')
plt.show()
```

## Равномерное распределение элементов и данных

```python
dfO = pd.read_csv('../csv/O11.csv', delimiter=';')
dfA = pd.read_csv('../csv/A11.csv', delimiter=';')
dfB = pd.read_csv('../csv/B11.csv', delimiter=';')
dfC = pd.read_csv('../csv/C11.csv', delimiter=';')

plt.figure(figsize=(10,6))

sns.regplot(data=dfO, x="n", y="t_ns", order=1, ci = 100, label = "O",  line_kws={'linestyle': '-'}, scatter=1)
sns.regplot(data=dfA, x="n", y="t_ns", order=1, ci = 100, label = "A",  line_kws={'linestyle': '-'}, scatter=1)
sns.regplot(data=dfB, x="n", y="t_ns", order=1, ci = 100, label = "B", line_kws={'linestyle': '-.'}, scatter=1)
sns.regplot(data=dfC, x="n", y="t_ns", order=1, ci = 100, label = "C",  line_kws={'linestyle': ':'}, scatter=1)
# кресты ошибок - доверительный интервал 95%

# plt.yscale('log')
# plt.xscale('log')

plt.title('Реализация find_ABC в сравнении с find_O\n равномерные данные и запросы \n Линейный масштаб')
plt.xlabel('$N$')
plt.ylabel('$\\tau,\\ \\text{ns}$')
plt.grid(True, alpha=0.3)
plt.legend()
# plt.ylim(0, 50)
# plt.xlim(0, 100000)
plt.savefig('abc_test_11.png')
plt.show()
```
